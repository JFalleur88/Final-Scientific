# Monte Carlo Sim Model

1. Winning percentage. 
2. Number of wins(one coach)$/$total wins(all coaches)
3. Number of losses $/$ total losses
4. More years of coaching = less chance to lose
5. More years of coaching = more points gain when wins

Right now, these factors will affect the resulf of this simulation. The first walk graph is without those factors. The sencond walk graph is using the factors above.

## Football Sim

```python
totalwinsfootball = 0
totallossesfootball = 0
totalyearsfootball = 0
for i in range(len(datafootball)):
    totalwinsfootball += int(datafootball[i][3])
    totallossesfootball += int(datafootball[i][4])
    totalyearsfootball += int(datafootball[i][2])
```

```python
trialsfootball = 1000
timefootball = 1000
initfootball = 100
overallfootball = []
for t in range(len(datafootball)):
    initialfootball = [initfootball]
    for i in range(timefootball):
        if np.random.random() < (float(datafootball[t][6])):
            initialfootball.append(initialfootball[-1] + 1)
        else:
            initialfootball.append(initialfootball[-1] - 1)
        if initialfootball[-1] == 0:
            break
    if len(initialfootball) < timefootball + 1:
        initialfootball += [0] * (timefootball + 1 - len(initialfootball))
    overallfootball.append(initialfootball)
```

```python
b = np.linspace(0, timefootball, timefootball + 1)
for o in overallfootball:
    plt.plot(b, o)
    plt.xlim(0, 1000)
    plt.title('Football Monte Carlo Sim with win ratio factor only')
    plt.xlabel('Time')
    plt.ylabel('Wins')
```

```python
trialsfootball = 1000
timefootball = 1000
initfootball = 100
overallfootball = []
for t in range(len(datafootball)):
    initialfootball = [initfootball]
    for i in range(timefootball):
        if np.random.random() < (float(datafootball[t][6]) + (float(datafootball[t][3]) / totalwinsfootball) - 
                                 (float(datafootball[t][4]) / totallossesfootball)
                                - ((65 - float(datafootball[t][2])) / totalyearsfootball)):
            initialfootball.append(initialfootball[-1] + 1 + float(datafootball[t][2]) * 0.01)
        else:
            initialfootball.append(initialfootball[-1] - 1)
        if initialfootball[-1] == 0:
            break
    if len(initialfootball) < timefootball + 1:
        initialfootball += [0] * (timefootball + 1 - len(initialfootball))
    overallfootball.append(initialfootball)
```

```python
b = np.linspace(0, timefootball, timefootball + 1)
for o in overallfootball:
    plt.plot(b, o)
    plt.title('Football Monte Carlo Sim with all factors')
    plt.xlabel('Time')
    plt.ylabel('Wins')
```
